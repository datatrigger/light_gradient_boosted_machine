```python
# Import necessary modules and data
import pandas as pd
import lightgbm as lgb
from random import seed
from time import time
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
import xgboost as xgb

df = pd.read_csv("train.csv")
```

```python
df.describe()
```

```python
#NAs
print(f'There are {df.isna().sum().sum()} missing values in the dataset.')
```

```python
# Train/Test split the data
seed(42)
train, test = train_test_split(df, test_size = 0.25)
```

```python
#LightGBM
train_data = lgb.Dataset(train[train.columns[:-1]], label=train[train.columns[-1:]])
test_x = test[test.columns[:-1]]
test_y = test[test.columns[-1:]]
```

```python
#LightGBM parameters based on the original paper https://arxiv.org/abs/1803.10260
parameters = {
    'boosting_type': 'gbdt',
    'objective': 'regression',
    'metric': 'l2',
    'num_leaves': 31,
    'learning_rate': 0.02,
    'max_depth' : 16,
    'min_data_in_leaf' : 1,
    'feature_fraction': 0.5,
    'bagging_fraction': 0.5,
    'bagging_freq': 1,
    'verbose': 1
}
```

```python
num_trees = 374

start = time()
lgbm = lgb.train(
    params = parameters,
    train_set = train_data,
    num_boost_round = num_trees,
    verbose_eval = 1
)
execution_time = time() - start
print(f'LightGBM execution time (374 iterations): {execution_time:0.2f} s')
```

```python
# Train the LightGBM model
start = time()
gbm = lgb.cv(params = parameters,
                train_set = train_data,
                #valid_sets = train_data,
                num_boost_round=374,
                #early_stopping_rounds=10,
                nfold = 5,
                stratified = False,
                verbose_eval = 1
            )
execution_time = time() - start
```

```python
print(f'LightGBM CV execution time: {execution_time:0.2f} s')
```

```python
gbm['l2-mean'].index(min(gbm['l2-mean']))
```

```python
min(gbm['l2-mean'])**0.5
```

```python
gbm_final = lgb.train(parameters, train_data, gbm['l2-mean'].index(min(gbm['l2-mean'])))
pred_y = gbm_final.predict(test_x)
```

```python
print(f'LightGBM RMSE on test sample: {mean_squared_error(test_y, pred_y) ** 0.5:0.2f} Â°K')
```

```python
#XGB
dtrain = xgb.DMatrix(train[train.columns[:-1]], label=train[train.columns[-1:]])
dtest = xgb.DMatrix(test_x, label = test_y)
```

```python
# specify parameters via map
parameters = {
    'booster': 'gbtree',
    'objective': 'reg:squarederror',
    'max_leaves': 31,
    'eta': 0.02,
    'max_depth' : 16,
    'min_child_weight' : 1,
    'colsample_bytree': 0.5,
    'subsample': 0.5
}
```

```python
start = time()
bst_cv = xgb.cv(
    params = parameters,
    dtrain = dtrain,
    num_boost_round=2000,
    early_stopping_rounds=10,
    nfold = 5,
    stratified = False,
    metrics = 'rmse'
)
execution_time = time() - start
```

```python
print(f'XGBoost CV execution time: {execution_time:0.2f} s')
```
