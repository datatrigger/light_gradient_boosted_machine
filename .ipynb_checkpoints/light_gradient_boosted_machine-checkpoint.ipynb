```python
import pandas as pd
import lightgbm as lgb

df = pd.read_csv("train.csv")
```

```python
df
```

```python
from sklearn.model_selection import train_test_split
train, test = train_test_split(df, test_size = 0.25)

train_data = lgb.Dataset(train[train.columns[:-1]], label=train[train.columns[-1:]])
test_x = test[test.columns[:-1]]
test_y = test[test.columns[-1:]]

validation_data = lgb.Dataset(test_x, test_y, reference = train_data)
```

```python
parameters = {
    'boosting_type': 'gbdt',
    'objective': 'regression',
    'metric': 'l2',
    #'num_iterations' : 374,
    'num_leaves': 31,
    'learning_rate': 0.02,
    'max_depth' : 16,
    'min_data_in_leaf' : 1,
    'feature_fraction': 0.9,
    'bagging_fraction': 0.5,
    'bagging_freq': 1,
    'verbose': 1
}
```

```python
gbm = lgb.train(params = parameters,
                train_set = train_data,
                valid_sets = validation_data,
                num_boost_round=1000,
               early_stopping_rounds=5)
```

```python
pred_y = gbm.predict(test_x, num_iteration=gbm.best_iteration)
```

```python
gbm.best_iteration
```

```python
from sklearn.metrics import mean_squared_error
print('The rmse of prediction is:', mean_squared_error(test_y, pred_y) ** 0.5)
```
